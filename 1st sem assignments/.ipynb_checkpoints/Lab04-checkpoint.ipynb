# **Question 1.(a)**

```python
def isPrime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    r = int(n**0.5)
    for d in range(3, r + 1, 2):
        if n % d == 0:
            return False
    return True

def prFun(i):
    bigP = i + 1
    smallP = i - 1

    # smallest prime > i
    while not isPrime(bigP):
        bigP += 1

    # largest prime < i
    while smallP > 1 and not isPrime(smallP):
        smallP -= 1
    if smallP <= 1:
        smallP = None   # no smaller prime exists (e.g., i <= 2)

    return smallP, bigP

# sample checks
print("prFun(20) =", prFun(20))  # (19, 23)
print("prFun(41) =", prFun(41))  # (37, 43)
print("prFun(2) =", prFun(2))  # (None, 3)
print("prFun(72) =", prFun(72))  # (71, 73)
print("prFun(100) =", prFun(100))  # (97, 101)
```

# **Question 2.(a)**

```python
import pyomo.environ as pyo

## Create an empty optimization model called 'm'
m = pyo.ConcreteModel()

## first add variables to the model 'm'
m.x1 = pyo.Var(within=pyo.NonNegativeReals)
m.x2 = pyo.Var(within=pyo.NonNegativeReals)
m.y1 = pyo.Var(within=pyo.Reals, bounds=(None,1000.0))

## Create an objective function
m.obj = pyo.Objective(sense=pyo.minimize, expr= (3*m.x1 + 4*m.x2 - 2*m.y1))

## Add constraints
m.con1 = pyo.Constraint(expr=(6*m.x1 + 8*m.x2 - 10*m.y1 >= 200))
m.con2 = pyo.Constraint(expr=(6*m.x2 + 5*m.y1 <= 16))

## Now create a solver. It needs a name and the location where it is installed
solver = pyo.SolverFactory("cbc",executable='/usr/bin/cbc')

## Solve using the selected solver.
result = solver.solve(m, tee=False)

## Show results.
print("Results: ")
print("Solver termination status:",result.solver.status)
print("Solver termination condition:", result.solver.termination_condition)
print(f"Solver time: {result.solver.time} seconds")
print("Solution value:", m.obj())

print()
print("Solution:")
print("x1 =",m.x1.value)
print("x2 =",m.x2.value)
print("y1 =",m.y1.value)

print()
print("Constraint Activity at Solution:")
print("con1:", m.con1())
print("con2:",m.con2())
```

```python
import pyomo.environ as pyo

## Create an empty optimization model called 'm'
m = pyo.ConcreteModel()

## first add variables to the model 'm'
m.x1 = pyo.Var(within=pyo.NonNegativeReals)
m.x2 = pyo.Var(within=pyo.NonNegativeReals)
m.y1 = pyo.Var(within=pyo.Reals, bounds=(10,1000.0))

## Create an objective function
m.obj = pyo.Objective(sense=pyo.minimize, expr= (3*m.x1 + 4*m.x2 - 2*m.y1))

## Add constraints
m.con1 = pyo.Constraint(expr=(6*m.x1 + 8*m.x2 - 10*m.y1 >= 200))
m.con2 = pyo.Constraint(expr=(6*m.x2 + 5*m.y1 <= 16))

## Now create a solver. It needs a name and the location where it is installed
solver = pyo.SolverFactory("cbc",executable='/usr/bin/cbc')

## Solve using the selected solver.
result = solver.solve(m, tee=False)

## Show results.
print("Results: ")
print("Solver termination status:",result.solver.status)
print("Solver termination condition:", result.solver.termination_condition)
print(f"Solver time: {result.solver.time} seconds")
print("----"*15)

try:
    print("Solution value:", m.obj())
    print()
    print("Solution:")
    print("x1 =",m.x1.value)
    print("x2 =",m.x2.value)
    print("y =",m.y1.value)

    print()
    print("Constraint Activity at Solution:")
    print("con1:", m.con1())
    print("con2:",m.con2())
except:
    print("----"*15)
    print("Given problem has no optimal solution")
```

# **Question 3 (all parts)**

```python
import pyomo.environ as pyo
from pyomo.environ import Var, NonNegativeReals, Reals, Objective, minimize, Constraint, SolverFactory

## Create an empty optimization model called 'm3'
m3 = pyo.ConcreteModel()

## Define variables with their bounds
m3.x1 = Var(within=Reals, bounds=(-1.0, None))
m3.x2 = Var(within=Reals, bounds=(3.0, None))
m3.x3 = Var(within=Reals, bounds=(-10.0, None))
m3.x4 = Var(within=Reals, bounds=(8.0, None))
m3.x5 = Var(within=Reals, bounds=(9.0, None))
m3.x6 = Var(within=Reals, bounds=(0.0, None))
m3.x7 = Var(within=Reals, bounds=(0.0, None))

## Create the objective function
expr_obj = 10*m3.x1 + 5*m3.x2 - 3*m3.x3 - 6*m3.x4 + 9*m3.x5 - 7*m3.x6 - 1.6*m3.x7
m3.obj = Objective(sense=minimize, expr=expr_obj)

## Add constraints
m3.con1 = Constraint(expr=(m3.x1 + m3.x2 + m3.x3 + m3.x4 + m3.x5 + m3.x6 + m3.x7 == 100))
m3.con2 = Constraint(expr=(m3.x1 + m3.x2 >= 10))
m3.con3 = Constraint(expr=(4*m3.x3 + 2*m3.x6 + 3*m3.x7 <= 150))
m3.con4 = Constraint(expr=(10*m3.x1 - m3.x6 + 3*m3.x5 >= 20))
m3.con5 = Constraint(expr=(m3.x3 - m3.x5 + 3*m3.x7 >= 10))


## Create a solver
solver = SolverFactory("cbc")

## Solve the model
result = solver.solve(m3, tee=False)

## Report the results
print("Results for Model 'm3':")
print("Solver status:", result.solver.status)
print("Solver termination condition:", result.solver.termination_condition)
print(f"Solver time: {result.solver.time} seconds")
print("----"*15)

try:
    print()
    print(f"Solution Value: {m3.obj()}")
    print()
    print("Optimal Variable Values:")
    print(f"x1 = {m3.x1.value:.4f}")
    print(f"x2 = {m3.x2.value:.4f}")
    print(f"x3 = {m3.x3.value:.4f}")
    print(f"x4 = {m3.x4.value:.4f}")
    print(f"x5 = {m3.x5.value:.4f}")
    print(f"x6 = {m3.x6.value:.4f}")
    print(f"x7 = {m3.x7.value:.4f}")
    print()
    print("Constraint Activity at Solution:")
    print(f"Constraint 1: {m3.con1()} (= 100)")
    print(f"Constraint 2: {m3.con2()} (>= 10)")
    print(f"Constraint 3: {m3.con3()} (<= 150)")
    print(f"Constraint 4: {m3.con4()} (>= 20)")
    print(f"Constraint 5: {m3.con5()} (>= 10)")
except:
    print("----"*15)
    print("Given problem has no optimal solution")

```

# **Question 4 (all parts)**

```python
import pyomo.environ as pyo
from pyomo.environ import Var, NonNegativeReals, Reals, Objective, maximize, Constraint, SolverFactory

## Create a new model
m = pyo.ConcreteModel()

## Define variables
m.x1 = Var(within=NonNegativeReals)
m.x2 = Var(within=NonNegativeReals)
m.x3 = Var(within=NonNegativeReals)
m.x4 = Var(within=NonNegativeReals)
m.x5 = Var(within=NonNegativeReals)
m.y = Var(within=Reals, bounds=(0.0, 10.0))

## Define the objective function
obj_expr = 0.043*m.x1 + 0.027*m.x2 + 0.025*m.x3 + 0.022*m.x4 + 0.045*m.x5 - 0.0275*m.y
m.obj = Objective(sense=maximize, expr=obj_expr)

## Define the constraints based on the reformulated LP
m.con1 = Constraint(expr=(m.x1 + m.x2 + m.x3 + m.x4 + m.x5 - m.y <= 100))
m.con2 = Constraint(expr=(m.x2 + m.x3 + m.x4 >= 40))
m.con3 = Constraint(expr=(0.6*m.x1 + 0.6*m.x2 - 0.4*m.x3 - 0.4*m.x4 + 3.6*m.x5 <= 0))
m.con4 = Constraint(expr=(4*m.x1 + 10*m.x2 - m.x3 - 2*m.x4 - 3*m.x5 <= 0))

## Create a solver and solve the model
solver = SolverFactory("cbc")
result = solver.solve(m, tee=False)

## Report the results
print("Results:")
print("Solver status:", result.solver.status)
print("Solver termination condition:", result.solver.termination_condition)
print(f"Solver time: {result.solver.time} seconds")
print("----"*15)

try:
    print()
    print(f"Solution Value: {m.obj()}")
    print()
    print("Optimal Variable Values:")
    print(f"x1 = {m.x1()}")
    print(f"x2 = {m.x2()}")
    print(f"x3 = {m.x3()}")
    print(f"x4 = {m.x4()}")
    print(f"x5 = {m.x5()}")
    print(f"y  = {m.y()}")
    print()
    print("Constraint Activity at Solution:")
    print(f"Constraint 1: {m.con1()} (<= 100)")
    print(f"Constraint 2: {m.con2()} (>= 40)")
    print(f"Constraint 3: {m.con3()} (<= 0)")
    print(f"Constraint 4: {m.con4()} (<= 0)")
except:
    print("----"*15)
    print("Given problem has no optimal solution")
```
